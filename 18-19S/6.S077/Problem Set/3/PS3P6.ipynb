```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[0.39, 0.72, 1.00, 1.52, 5.20, 9.58, 19.20, 30.05, 39.48]]).T
Y = np.array([[0.24, 0.615, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.1]]).T
```

```python
plt.scatter(X, Y)
```

```python
plt.scatter(np.log(X), np.log(Y))
```

```python
ols = LinearRegression()
ols.fit(np.log(X), np.log(Y))
```

```python
plt.plot(np.arange(-2, 4, 0.01), ols.predict(np.arange(-2, 4, 0.01).reshape(-1, 1)))
plt.scatter(np.log(X), np.log(Y))
```

```python
plt.plot(np.exp(np.arange(-2, 4, 0.01)), np.exp(ols.predict(np.arange(-2, 4, 0.01).reshape(-1, 1))))
plt.scatter(X, Y)
```

```python
ols.coef_
```

```python
np.exp(ols.intercept_)
```

```python
plt.plot(np.arange(0, 50, 0.01), (np.exp(ols.intercept_)*np.arange(0, 50, 0.01)**ols.coef_).reshape(-1))
plt.scatter(X, Y)
```

```python
np.sqrt(np.std(np.log(X)))
```

```python
R_square = ols.score(np.log(X), np.log(Y))
```

```python
F = R_square/(1 - R_square)*((9 - 1 - 1)/1)
```
