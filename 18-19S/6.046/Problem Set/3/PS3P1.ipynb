### Part A

```python
n = 2 * np.random.randint(10, 25 + 1) + 1
origin = n // 2

crumb = np.random.randint(0, n)
while crumb == origin:
    crumb = np.random.randint(0, n)
    
log = np.zeros(n)
log[crumb] = 1
optimal_distance = abs(crumb - origin)
```

```python
found_crumb = False
origin_relative_location = 0
distance_traveled = 0

while not found_crumb:
    
    origin_relative_location += 1

    if origin + origin_relative_location == crumb:
        
        found_crumb = True
        
        distance_traveled += origin_relative_location
        
    if origin - origin_relative_location == crumb:
        
        found_crumb = True
        
        distance_traveled += 2 * origin_relative_location
    
    if not found_crumb:
        distance_traveled += origin_relative_location
```
